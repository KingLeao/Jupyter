#  Generating Power-Law Distribution Random Numbers

```python
import random
import matplotlib.pyplot as plt
import time
from pandas import Series
beginTime = time.clock()

# Acceptance-Rejection Method
# time-consuming
def powerlaw_rnd(v,xmin=0.01):
    C = (v-1)/(xmin**(1-v)-1)
    while True:
        x = random.uniform(xmin,1.0)
        y = random.uniform(1,xmin**(-v))
        if y <= x**(-v):
            return x

# Inverse Transform Method
# more effecient
def powerlaw_num(v,xmin=0.01):
    x = random.random()
    return (xmin**(1-v)-(xmin**(1-v)-1)*x)**(1./(1-v))

y = []
for i in range(10000):
#     num = powerlaw_rnd(3,0.01)
    num = powerlaw_num(3,0.01)
    y.append(num)
    
print sum(y)/float(len(y))
# print y
L = Series(y)
L.hist(bins=100,alpha=0.3,color='k',normed=True) 
L.plot(kind='kde')

plt.show()

endTime = time.clock()
print "Running time: %s Seconds" % (endTime - beginTime)
```

```python
# -*- coding: utf-8 -*-
import random
import networkx as nx
import matplotlib.pyplot as plt

def init_activityDriven(N=1000, m=4):
    """"
    Parameters
    ----------
    N :   nodes
    m :   rewiring links
    """
    # initiate parameters
    # N = 100   # nodes
    # m = 4     # rewiring links
    dt = 1    # active slot
    eta = 20  # rescaling factor , ai = eta * xi

    G = nx.Graph()
    G.add_nodes_from(range(N))
    for i in range(N):
        xi = powerlaw_num(3, 0.01)  # node's activated prob belongs to F(x)=x**(-3)
        if random.random() <= eta * xi * dt:
            for mi in range(m):
                j = random.choice(range(N))
                while j == i or G.has_edge(i, j):
                    j = random.choice(range(N))
                G.add_edge(i, j)
    return G  # return a graph object

# Generating T time steps time-varying activity driven networks
def temporal_activityDriven(N=1000, m=4, T=6):
    """"
    Parameters
    ----------
    N :   nodes
    m :   rewiring links
    T :   time slots
    """
    tG = {}
    for i in range(1, T+1):
        G_temp = init_activityDriven(N, m)
        tempNet = nx.to_dict_of_dicts(G_temp)
        tG[i] = tempNet
    return tG  # return a dict-like time-varying graph

G = temporal_activityDriven(1000,100,6)
for i in range(1,7):
    G_temp = nx.Graph(G[i])
    print nx.density(G_temp)
    print nx.is_connected(G_temp)
#     G.append(G_temp)

    nx.draw(G_temp, pos=nx.spring_layout(G_temp))

    plt.show()
```

# Generating Acitivity Driven Time-Varying Networks with Memory Mechanism

```python
# -*- coding: utf-8 -*-
import random
import networkx as nx
import matplotlib.pyplot as plt

# Inverse Transform Method
# more efficient
def powerlaw_num(v,xmin=0.01):
    x = random.random()
    return (xmin**(1-v)-(xmin**(1-v)-1)*x)**(1./(1-v))

# Generating an empty graph
def emptyGraph(N=1000):
    G = nx.Graph()
    G.add_nodes_from(range(N))
    return G

# Generating activity driven time-varying network in one timeslot
def init_activityDriven(N=1000, m=6):
    """"
    Parameters
    ----------
    N :   nodes
    m :   rewiring links
    """
    # initiate parameters
    # N = 1000   # nodes
    # m = 6     # rewiring links
    dt = 1    # active slot
    eta = 10  # rescaling factor , ai = eta * xi

    G = nx.Graph()
    G.add_nodes_from(range(N))
    for i in range(N):
        xi = powerlaw_num(3, 0.03)  # node's activated prob belongs to F(x)=x**(-3)
        if random.random() <= eta * xi * dt:
            for mi in range(m):
                j = random.choice(range(N))
                while j == i or G.has_edge(i, j):
                    j = random.choice(range(N))
                G.add_edge(i, j)
    return G  # return a graph object

# Generating T time steps time-varying activity driven networks
def temporal_activityDriven(N=1000, m=6, T=10):
    """"
    Parameters
    ----------
    N :   nodes
    m :   rewiring links
    T :   time slots
    """
    tG = {}
    for i in range(1, T+1):
        G_temp = init_activityDriven(N, m)
        tempNet = nx.to_dict_of_dicts(G_temp)
        tG[i] = tempNet
    return tG  # return a dict-like time-varying graph


def temporal_activityDriven_withMemory(N=1000, m=6, T=10, r=0):
    """"
    Parameters
    ----------
    N :   nodes
    m :   rewiring links
    T :   time slots
    r :   memory rate, keeping previous links
    """
    # initiate parameters
    # N = 1000   # nodes
    # T = 1000  # time slots
    # r = 0.1   # memory rate, keeping previous links
    dt = 1    # active slot
    eta = 10  # rescaling factor , ai = eta * xi

    G0 = init_activityDriven(N, m)  # initial graph
    G1 = G0   # current timeslot graph
    tG = {}   # graph dicts
    for ti in range(T):
        G2 = G1  # previous timeslot graph
        G1 = G0  # current timeslot graph
        for i in range(N):
            xi = powerlaw_num(3, 0.03)  # node's activated prob belongs to F(x)=x**(-3)
#             print xi*eta
            if random.random() < eta * xi * dt:  # node i is active
                # ki = nx.degree(G2, i)            # calculate degree of node i
                neighbors = G2.neighbors(i)        # neighbors list of node i
                # link selection of each neighbor
                for neighbor in neighbors:
                    if random.random() < r:
                        # randomly keep one previous neighbor
                        j = neighbor
                    else:
                        # randomly select one other node
                        G1.remove_edge(i,neighbor)
                        j = random.choice(list(set(range(N)) - {i} - set(neighbors)))
                        count = 0
                        while G1.has_edge(i, j):
                            j = random.choice(list(set(range(N)) - {i} - set(neighbors)))
                            count += 1
                    G1.add_edge(i, j)
                    
        tempNet = nx.to_dict_of_dicts(G1)
        tG[ti] = tempNet
    return tG  # return a dict-like time-varying graph

# # tG = temporal_activityDriven_withMemory(1000,10,6,0.2)
# tG = temporal_activityDriven(1000,100,6)
# for i in range(1,7):
#     tempNet = nx.Graph(tG[i])
#     print nx.density(tempNet)
    
#     degree = nx.degree_histogram(tempNet)
#     x = range(len(degree))
#     y = [z/float(sum(degree)) for z in degree]
    
#     plt.plot(x,y,'r-*')
# #     nx.draw(tempNet,pos=nx.random_layout(tempNet))
#     plt.show()

#############################################################################
### this is the simplest method to construc AD networks
### always with same amount of links
### use prob to decide whether keep the active links or not
#############################################################################
    
# activity driven networks for one time step
# graph = init_activityDriven(1000,4)
# tG = temporal_activityDriven(N=1000, m=1, T=3)
tG = temporal_activityDriven_withMemory(N=1000, m=2, T=3, r=0.2)


for i in range(1,2):
    graph = nx.Graph(tG[i])
    print nx.density(graph)
    degree = nx.degree_histogram(graph)
    x = range(len(degree))
    y = [z/float(sum(degree)) for z in degree]

    plt.loglog(x,y,'g-o')
#     nx.draw(tempNet,pos=nx.random_layout(tempNet))
plt.show()
```

```python
# -*- coding: utf-8 -*-
import random
import networkx as nx
import matplotlib.pyplot as plt

# Inverse Transform Method
# more effecient
def powerlaw_num(v,xmin=0.01):
    x = random.random()
    return (xmin**(1-v)-(xmin**(1-v)-1)*x)**(1./(1-v))

# Generating an empty graph
def emptyGraph(N):
    G = nx.Graph()
    G.add_nodes_from(range(N))
    return G

# Generating activity driven time-varying network in one timeslot
def init_activityDriven(N=1000, m=6):
    """"
    Parameters
    ----------
    N :   nodes
    m :   rewiring links
    """
    # initiate parameters
    # N = 1000   # nodes
    # m = 6     # rewiring links
    dt = 1    # active slot
    eta = 20  # rescaling factor , ai = eta * xi

    G = nx.Graph()
    G.add_nodes_from(range(N))
    for i in range(N):
        xi = powerlaw_num(3, 0.01)  # node's activated prob belongs to F(x)=x**(-3)
        if random.random() <= eta * xi * dt:
            for mi in range(m):
                j = random.choice(range(N))
                while j == i or G.has_edge(i, j):
                    j = random.choice(range(N))
                G.add_edge(i, j)
    return G  # return a graph object

# Generating T time steps time-varying activity driven networks
def temporal_activityDriven(N=1000, m=6, T=10):
    """"
    Parameters
    ----------
    N :   nodes
    m :   rewiring links
    T :   time slots
    """
    tG = {}
    for i in range(1, T+1):
        G_temp = init_activityDriven(N, m)
        tempNet = nx.to_dict_of_dicts(G_temp)
        tG[i] = tempNet
    return tG  # return a dict-like time-varying graph


def temporal_activityDriven_withMemory(N=1000, m=6, T=5, r=0):
    """"
    Parameters
    ----------
    N :   nodes
    m :   rewiring links
    T :   time slots
    r :   memory rate, keeping previous links
    """
    # initiate parameters
    # N = 100   # nodes
    # T = 1000  # time slots
    # r = 0.1   # memory rate, keeping previous links
    dt = 1    # active slot
    eta = 20  # rescaling factor , ai = eta * xi

    G0 = init_activityDriven(N, m)  # initial graph
    G1 = G0   # current timeslot graph
    tG = {1: nx.to_dict_of_dicts(G0)}  # graph dicts
    for ti in range(1, T+1):
        G2 = G1  # previous timeslot graph
        G1 = emptyGraph(N)  # current timeslot graph
        for i in range(N):
#             xi = powerlaw_num(3, 0.01)  # node's activated prob belongs to F(x)=x**(-3)
            xi = 0.02
#             print xi*eta
            if random.random() < eta * xi * dt:  # node i is active
#                 ki = nx.degree(G2, i)            # calculate degree of node i
                neighbors = G2.neighbors(i)      # neighbors list of node i
                # link selection of each neighbor
                if len(neighbors) > 0:
                    for neighbor in neighbors:
                        if random.random() < r:
                            # randomly keep one previous neighbor
                            j = neighbor
                            G1.add_edge(i, j)
                        else:
                            # randomly select one other node
                            # G1.remove_edge(i,neighbor)
                            j = random.choice(list(set(range(N)) - {i} - set(neighbors)))
                            count = 0
                            while G1.has_edge(i, j):
                                j = random.choice(list(set(range(N)) - {i} - set(neighbors)))
                                count += 1
                            G1.add_edge(i, j)
                else:
                    for mi in range(m):
                        j = random.choice(range(N))
                        while j == i or G1.has_edge(i, j):
                            j = random.choice(range(N))
                        G1.add_edge(i, j)
                    
        tempNet = nx.to_dict_of_dicts(G1)
        tG[ti] = tempNet
    return tG  # return a dict-like time-varying graph


N = 5000
m = 10
T = 10
p = 1.0

tG = temporal_activityDriven_withMemory(N, m, T, p)
# tG = temporal_activityDriven(1000, 6, 100)


for i in range(1,10):
    fig=plt.figure()
    ax=fig.add_subplot(1,1,1)

    graph = nx.Graph(tG[i])
    print 'density:',nx.density(graph)
    print 'nodes:',len(graph)
    print "edges:",nx.number_of_edges(graph)
    degree = nx.degree_histogram(graph)
    x = range(len(degree))
    y = [z/float(sum(degree)) for z in degree]

    ax.set_xlabel("k",fontsize=15)
    ax.set_ylabel("P(k)",fontsize=15)
    ax.set_title("degree distribution",fontsize=15)
    plt.plot(x,y,'g-o')

# figpath = "F:\\PAPER_FIG\\AD_N_%s_m_%s_p_%s_PDF.png" % (N, m, p)
# plt.savefig(figpath)

plt.show()

# for i in range(1,201):
#     tempNet = nx.Graph(tG[i])
#     print nx.density(tempNet)
#     nx.draw(tempNet,pos=nx.random_layout(tempNet))
#     plt.show()
```

## Generating Acitivity Driven Time-Varying Networks with Memory Mechanism and identical activity potential 

```python
# -*- coding: utf-8 -*-
import random
import networkx as nx
import matplotlib.pyplot as plt
        
# Generating an empty graph
def emptyGraph(N):
    G = nx.Graph()
    G.add_nodes_from(range(N))
    return G

# Generating activity driven time-varying network in one timeslot
# All nodes are activated as ER random network ER(N,p), where p = m/N
def init_activityDriven(N=1000, m=4):
    """"
    Parameters
    ----------
    N :   nodes
    m :   rewiring links
    """
    # initiate parameters
    # N = 1000   # nodes
    # m = 6     # rewiring links
    dt = 1    # active slot
    eta = 10  # rescaling factor , ai = eta * xi

    G = nx.Graph()
    G.add_nodes_from(range(N))
    for i in range(N):
        xi = powerlaw_num(3, 0.01)  # node's activated prob belongs to F(x)=x**(-3)
#         xi = 0.1
        if random.random() <= eta * xi * dt:
            for mi in range(m):
                j = random.choice(range(N))
                while j == i or G.has_edge(i, j):
                    j = random.choice(range(N))
                G.add_edge(i, j)
    return G  # return a graph object

def temporal_activityDriven_withMemory(N=1000, m=3, T=5, r=0):
    """"
    Parameters
    ----------
    N :   nodes
    m :   rewiring links
    T :   time slots
    r :   memory rate, keeping previous links
    """
    # initiate parameters
    # N = 1000  # nodes
    # T = 1000  # time slots
    # r = 0.1   # memory rate, keeping previous links
    
    G0 = init_activityDriven(N, m)  # initial graph
    G1 = G0             # current timeslot graph
    tG = {1:nx.to_dict_of_dicts(G0)}        # graph dicts
    for ti in range(2, T+1):
        G2 = G1            # previous timeslot graph
        G1 = emptyGraph(N) # current timeslot graph
        for i in range(N):
            neighbors = G2.neighbors(i)     # neighbors list of node i
            # link selection on each neighbor
            for neighbor in neighbors:
                if random.random() < r:
                    # reserve link with previous neighbor
                    j = neighbor
                    G1.add_edge(i, j)
                else:
                    # randomly select a new node
                    j = random.choice(list(set(range(N)) - {i} - set(neighbors)))
                    count = 0
                    while G1.has_edge(i, j):
                        j = random.choice(list(set(range(N)) - {i} - set(neighbors)))
                        count += 1
                    G1.add_edge(i, j)
                    
        tempNet = nx.to_dict_of_dicts(G1)
        tG[ti] = tempNet
    return tG  # return a dict-like time-varying graph

N = 1000
m = 5
T = 6
r = 0.8

tG = temporal_activityDriven_withMemory(N, m, T, r)

# draw degree distribution
for i in range(1,7):
    fig = plt.figure()
    ax=fig.add_subplot(1,1,1)

    graph = nx.Graph(tG[i])
    print nx.density(graph)
    print nx.number_of_edges(graph)
    degree = nx.degree_histogram(graph)
    x = range(len(degree))
    y = [z/float(sum(degree)) for z in degree]

    ax.set_xlabel("k",fontsize=15)
    ax.set_ylabel("P(k)",fontsize=15)
    ax.set_title("degree distribution",fontsize=15)
    plt.plot(x,y,'g-o')

# figpath = "F:\\PAPER_FIG\\AD_N_%s_m_%s_p_%s_idePDF.png" % (N, m, p)
# plt.savefig(figpath)

plt.show()

# for i in range(1,201):
#     tempNet = nx.Graph(tG[i])
#     print nx.density(tempNet)
#     nx.draw(tempNet,pos=nx.random_layout(tempNet))
#     plt.show()
```
