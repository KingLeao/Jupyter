```python
def runningMean(x, N):
    y = np.zeros((len(x),))
    for i in range(len(x)-3):
         y[i] = np.sum(x[i:(i+N)])/N
    y[-3]=(x[-3]+x[-2])/2
    y[-2]=(x[-2]+x[-1])/2
    y[-1]=x[-1]
    return y
```

# Draw Epidemic Threshold with vary keepN (AD)¶

```python
import numpy as np
import json
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# datapath = r"F:\simulationDATA\ActivityDriven_VaryS\DEMO"
datapath = r"F:\serverDATA\AD"

# beta = 0.2
gama = 0.4
simulationT = 100
p = 0.6

fig = plt.figure(figsize = (6,4))
ax = plt.subplot(111)

X = range(1, 11)

threshold = []
for keepN in range(1,11):
    prevalenceList = []
    for beta in np.arange(0.08,0.404,0.004):

        spreadingContent = json.load(open(datapath + "\\beta_%s_gama_%s_p_%s_keepN_%s_Average_test%s.json" \
                                             % (beta, gama, p, keepN, simulationT)))
        steadyList = spreadingContent[900:1000]
        if sum(steadyList)/100 > 0.01:
            prevalenceList.append(beta/gama)
    threshold.append(prevalenceList[0])  
# threshold = runningMean(threshold,4)
print "threshold\n",threshold    
ax.plot(X,threshold,'ro',markersize=10)

# ax.set_zlim(0,1.0)
ax.set_xlabel("keepN",fontsize=18)
ax.set_ylabel("threshold",fontsize=20)
ax.set_title("p = %s"%p,fontsize=15)
ax.set_xticks(np.arange(1,11,1))

# ax.set_xticks(np.arange(0.0,1.1,0.1))
ax.set_yticks(np.arange(0.26,0.36,0.02))

# save fig
# figpath = "F:\\FIG\\AD_threshold_beta_%s_gama_%s_p_%s.png"%(beta,gama,p)
# plt.savefig(figpath)
plt.show()
```
